```python
from pymongo import MongoClient

client = MongoClient('localhost', 27017)
db = client['task-1']
collection = db['world-facts']
```

```python
pipeline = [
    {"$project": {"product_name":1, "energy_100g":1, "countries_en":1, "brands":1}},
    # Get
    {"$match": {"product_name": {"$regex":".*peanut.butter.*", 
                                 "$options": "i"},
                "countries_en": {"$regex":".*australia.*",
                                  "$options": "i"}}},
    # Short
    {"$sort":{"energy_100g":-1}},
    # Gi
    {"$limit": 5},
]

pb_df = pd.DataFrame(list(collection.aggregate(pipeline)))
pb_df.head()
```

```python
import seaborn as sns

pb_df['brand'] = ["Coles - Crunchy", "The Forage Company - Smooth", "Sanitarium - Crunchy",
                  "Bramwells - Crunchy", "Bramwells - American Style"]
fig = sns.barplot(y="brand", x="energy_100g", data=pb_df, fill="brands", orient="h")
fig.set_title('Top 5 Energy Rich Peanut Butters in Australia', fontsize=12)
fig.set(xlabel="Ener", ylabel="pean")
```

```python
pipeline = [
    # SELECT Zinc, Countries 
    {"$project": {"zinc_100g":1, 'countries_en':1}},
   
    # WHERE zinc_100g > 0.001
    {"$match": {"zinc_100g": {"$gt": 0.001}}},

    # GROUP BY countries_en
    {"$group": {"_id": "$countries_en",
                "count":{"$sum": 1}}},
    
    # HAVING count(*) > 1
    {"$match": {"count": {"$gt": 1}}},
    
    # Sort by Energy Content
    {"$sort":{"count":-1}},

    # Take the first 5
    {"$limit": 10}
]
pb_df = pd.DataFrame(list(collection.aggregate(pipeline)))
pb_df.head(10)
```

```python
pipeline = [
    # SELECT Zinc, Countries 
    {"$project": {"zinc_100g":1, 'countries_en':1}},
   
    # WHERE zinc_100g > 0.001
    {"$match": {"zinc_100g": {"$gt": 0.001}}},

    # GROUP BY countries_en
    {"$group": {"_id": "$countries_en",
                "count":{"$sum": 1}}},
    
    # HAVING count(*) > 1
    {"$match": {"count": {"$gt": 1}}},
    
    # Sort by Energy Content
    {"$sort":{"count":-1}},

    # Take the first 5
    {"$limit": 10}
]
pb_df = pd.DataFrame(list(collection.aggregate(pipeline)))
pb_df.head(10)
from tabulate import tabulate
pb_df.rename(index=str, columns={"_id":"Country"}, inplace=True)
print(tabulate(pb_df, headers='keys', tablefmt='psql'))
```

```python
pipeline = [
    {"$project": {"categories_en": 1, "proteins_100g":1}}, 
    {"$match": {"categories_en": {"$regex":"(pork|chicken|tofu)", 
                                 "$options": "i"}}},
]
pb_df = pd.DataFrame(list(collection.aggregate(pipeline)))
pb_df["food_group"] = pb_df.categories_en.str.extract("(pork|chicken|tofu)")
nn_df = pb_df[pb_df.food_group.notnull()]
nn_df.groupby(['food_group']).agg(['mean', 'median', 'std'])
```
